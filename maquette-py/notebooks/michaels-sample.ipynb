```python
import maquette as mq
import pandas as pd
import pandavro as pdx
```

```python
def import_excel(file, dataset, title, summary, classification='public'):
    mq.Dataset(dataset, title=title, summary=summary).create()
    df = pd.read_excel('/Users/michaelwellner/Downloads/b2b_large.xlsx', engine='openpyxl', keep_default_na=False)
    df.columns = map(lambda s: ''.join(x for x in s.lower() if x.isalnum()), df.columns)
    mq.dataset(dataset).put(df, "Exported from D&B")
```
